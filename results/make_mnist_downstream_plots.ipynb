```python
import seaborn as sns; sns.set()
import pandas as pd
```

```python
# import matplotlib.style as style
# style.use('seaborn-whitegrid')
```

## Read data

```python
data = pd.read_csv('results_mnist_downstream.csv')
data.head()
```

```python
data_mar = data.loc[data["Mechanism"] == "MAR"]
data_mcar = data.loc[data["Mechanism"] == "MCAR"]

data_mcar = data_mcar[data_mcar["Accuracy"] > 0.5] # drop outlier
```

## Plotting

```python
plot = sns.lineplot(x="Missing Ratio", y="Accuracy", hue="Approach", markers = True, dashes = False, data = data_mar)
plot.get_figure().savefig('MNIST_Acc_MAR.pdf')
```

```python
plot = sns.lineplot(x="Missing Ratio", y="Accuracy", hue="Approach", markers = True, dashes = False, data = data_mcar)
plot.get_figure().savefig('MNIST_Acc_MCAR.pdf')
```
