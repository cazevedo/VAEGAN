```python
import seaborn as sns; sns.set()
import pandas as pd
```

```python
df = pd.read_csv("results_credit_rmse.csv")
df.replace("GAINCredit", 'GAIN', inplace=True)

df.columns = ["Approach", "Missing Data Mechanism", "Missing Ratio", 'NRMSE']

df = df.loc[df["Missing Data Mechanism"]=="MCAR"]
```

```python
df.head()
```

```python
ax = sns.catplot(x="Missing Ratio", y='NRMSE', hue='Approach', hue_order=['GAIN', 'MICE', 'MissForest', 'MeanImputation', 'MostFrequent'], data=df, kind='box')
ax.title('MCAR')
ax.savefig('Credit_NRMSE.pdf')
```

```python
ax = sns.catplot(x="Missing Ratio", y='R2', hue='Approach', hue_order=['GAIN', 'MICE', 'MissForest', 'MeanImputation', 'MostFrequent'],col='Missing Data Mechanism', data=df, kind='box')
ax.savefig('Credit_R2.pdf')
```

```python
ax = sns.catplot(x="Missing Ratio", y='MSE', hue='Approach', hue_order=['GAIN', 'MICE', 'MissForest', 'MeanImputation', 'MostFrequent'],col='Missing Data Mechanism', data=df, kind='box')
ax.savefig('Credit_MSE.pdf')
```
