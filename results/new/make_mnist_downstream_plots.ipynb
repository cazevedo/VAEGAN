```python
import seaborn as sns; sns.set()
import pandas as pd
```

## Read data

```python
data = pd.read_csv('results_mnist_downstream.csv')
data.head()
```

```python
data_mar = data.loc[data["Mechanism"] == "MAR"]
data_mcar = data.loc[data["Mechanism"] == "MCAR"]

data_mcar = data_mcar[data_mcar["Accuracy"] > 0.5] # drop outlier
```

## Plotting

```python
plot = sns.lineplot(x="Missing Ratio", y="Accuracy", hue="Approach", style="Approach", markers = True, dashes = True, data = data_mar)
plot.set_title('MAR')
fig = plot.get_figure().savefig('MNIST_Acc_MAR.pdf', bbox_inches='tight')
```

```python
plot = sns.lineplot(x="Missing Ratio", y="Accuracy", hue="Approach", style="Approach", markers = True, dashes = True, data = data_mcar)
plot.set_title('MCAR')
plot.get_figure().savefig('MNIST_Acc_MCAR.pdf', bbox_inches='tight')
```
